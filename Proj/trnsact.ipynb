# Data Cleaning Code. 
## Developed on 10,000 entries

```python
import pandas as pd
import numpy as np

dat = pd.read_csv('/home/henry/Desktop/MSiA/MSiA 400/MSiA400/Data/trnsact_10k.csv')
dat.head()
```

```python
percent_purchase = dat.groupby(['sku', 'stype'], group_keys=False)["stype"].aggregate(['count']).reset_index()
long = percent_purchase.pivot(index = 'sku', columns = 'stype', values = 'count').reset_index().fillna(0)
long["Percentage Purchase"] = long.apply(lambda x: 100 * x.P / (x.P + x.R), axis = 1)
long["Percentage Return"] = long.apply(lambda x: 100 * x.P / (x.P + x.R), axis = 1)


long
```
